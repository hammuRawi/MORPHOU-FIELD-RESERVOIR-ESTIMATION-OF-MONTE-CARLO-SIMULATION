   **INTRODUCTION**

    *****       MONTE CARLO SIMULATION  FOR RESERVE ESTIMATIONS    *****

 Monte Carlo Simulation belongs to the Monte Carlo Method family. The Monte Carlo Method is an experimental method that aims to reach the result by experimenting with random numbers. In this way, it is aimed to solve mathematical and physical problems. We could use exact variables which based on a fixed average value for each property that are the best estimates (typically values that are thought to be most likely) if we consider for deterministic approach. However,reserve estimates are not unique and can hardly be considered deterministic. Probabilistic approach is more realistic because of the uncertainties associated with reservoir characteristics, especially at locations without enough number of wells (without hard data).Therefore, probabilistic approach helps to estimate reserves accurately which based on a range of values (maximum/minimum and type of distribution) for each property, taking into account uncertainty in data.

```python
# WATER SATURATION VARIABLES
sw_mean = 0.11
sw_lower = 0.09              # lower limit
sw_upper = 0.13              # Upper limit
srn=np.random.random_sample(1000)
swr=(sw_mean-sw_lower)/(sw_upper-sw_lower)
sw_xn=[sw_lower+(swr*srn[i])**(1/2) if srn[i] < swr else sw_upper-((sw_upper-sw_mean)*(sw_upper-sw_lower)*(1-srn[i]))**(1/2)  for i in range(0,1000)]
```

```python
print(np.median(sw_xn))
print(np.average(sw_xn))
print(np.std(sw_xn))
```

```python
# NET-TO-GROSS (NTG) VARIABLES
ntg_mean = 0.75
ntg_min = 0.7               # scalar - Lower bound of the distribution support
ntg_max = 0.80               # scalar - Upper bound of the distribution support 
ntg_rn=np.random.random_sample(1000)
ntg_xr=(ntg_mean-ntg_min)/(ntg_max-ntg_min)
ntg_xn=[ntg_min+(ntg_xr*ntg_rn[i])**(1/2) if ntg_rn[i] < ntg_xr else ntg_max-((ntg_max-ntg_mean)*(ntg_max-ntg_min)*(1-ntg_rn[i]))**(1/2)  for i in range(0,1000)]
```

```python
print(np.median(ntg_xn))
print(np.average(ntg_xn))
print(np.std(ntg_xn))
```

```python
# WATER OIL CONTACT (WOC) VARIABLES
woc_mean = 11100
woc_min = 11000               # scalar - Lower bound of the distribution support
woc_max = 11200               # scalar - Upper bound of the distribution support 
woc_rn=np.random.random_sample(1000)
woc_xr=(woc_mean-woc_min)/(woc_max-woc_min)
woc_xn=[woc_min+(woc_xr*woc_rn[i])**(1/2) if woc_rn[i] < woc_xr else woc_max-((woc_max-woc_mean)*(woc_max-woc_min)*(1-woc_rn[i]))**(1/2)  for i in range(0,1000)]
```

```python
print(np.median(woc_xn))
print(np.average(woc_xn))
print(np.std(woc_xn))
```

```python
# GENERAL INPUT VARIABLES

n = 1000                   # number of samples or runs
seed = 121                   # fix the seed point - use RANDOM_STATE
bins = 500                   # number of bins

# SPECIFIC INPUT VARIABLES

#1. AREA (meters^2)          TRIANGULAR DISTRIBUTION
a_min=area[9]              # scalar - Lower bound of the distribution support
a_max=max(area)            # scalar - Upper bound of the distribution support
c1 = 0.5                     # scalar - The location of the triangle's peak 

#2. THICKNESS (meters)      TRIANGULAR DISTRIBUTION
t_min=min(thick)                  # scalar - Lower bound of the distribution support
t_max=max(thick)                 # scalar - Upper bound of the distribution support 
c2 = 0.5                     # scalar - The location of the triangle's peak 

#3. NET-TO-GROSS (NTG)      TRIANGULAR DISTRIBUTION
ntg_min = 0.7               # scalar - Lower bound of the distribution support
ntg_max = 0.80               # scalar - Upper bound of the distribution support 
c3 = 0.5                     # scalar - The location of the triangle's peak

#4. POROSITY                 TRUNCATED NORMAL DISTRIBUTION
por_mean=np.mean(por)                # Mean
por_std=np.std(por)                 # Standard Deviation
por_lower=min(por)            # lower limit
por_upper=max(por)           # Upper limit

#5. WATER SATURATION         TRUNCATED NORMAL DISTRIBUTION
sw_mean = 0.11               # Mean
sw_std = np.std(sw_xn)               # Standard Deviation
sw_lower = 0.09              # lower limit
sw_upper = 0.13              # Upper limit

#6. OIL FORMATION VOLUME FACTOR (Bo)   UNIFORM DISTRIBUTION
bo_min = 1.82                # scalar - Lower bound of the distribution support
bo_max = 1.85                # scalar - Upper bound of the distribution support

#7. WATER OIL CONTACT (WOC) VARIABLES
woc_mean = 11100
woc_std=np.std(woc_xn)
woc_min = 11000               # scalar - Lower bound of the distribution support
woc_max = 11200               # scalar - Upper bound of the distribution support 
```

```python
# Calculate Individual Probabilty Distributions using SciPy algorithms 
Area = stats.triang.rvs(c1, loc=a_min, scale=a_max-a_min, size=n, random_state = seed)
Thickness = stats.triang.rvs(c2, loc=t_min, scale=t_max-t_min, size=n, random_state = seed)
NTG = stats.triang.rvs(c3, loc=ntg_min, scale=ntg_max-ntg_min, size=n, random_state=seed)
# POR = stats.norm(por_mean, por_std).rvs(size = n, random_state=seed) 
POR = stats.truncnorm((por_lower - por_mean) / por_std, (por_upper - por_mean) / por_std, loc=por_mean, scale=por_std).rvs(size = n, random_state=seed) 
Sw = stats.truncnorm((sw_lower - sw_mean) / sw_std, (sw_upper - sw_mean) / sw_std, loc=sw_mean, scale=sw_std).rvs(size = n, random_state=seed) 
# Sw = stats.norm(sw_mean, sw_std).rvs(size = n, random_state=seed)
Bo = stats.uniform.rvs(loc = bo_min, scale=bo_max-bo_min, size=n, random_state = seed)
WOC= stats.uniform.rvs(loc = woc_min, scale=woc_max-woc_min, size=n, random_state = seed)
COUNTER=stats.uniform.rvs(loc = min(counter), scale=max(counter)-min(counter), size=n, random_state = seed)
# Plot Individual Probability Distributions to visualise inputs
# Set the figures
sns.set(color_codes=True)
sns.set_style("dark")
sns.set_style("ticks")
f, axes = plt.subplots(2,4, figsize=(16, 10), )
f.subplots_adjust(hspace=0.25, wspace=0.15)
# Seaborn distribution plots
sns.distplot( Area/1e6, color="skyblue", ax=axes[0, 0], axlabel='Area (MMft2)')
sns.distplot( Thickness , color="red", ax=axes[0, 1], axlabel='Thickness (ft)')
sns.distplot( NTG , color="gold", ax=axes[0, 2], axlabel='Net-to-Gross')
sns.distplot( POR , color="brown", ax=axes[0, 3], axlabel='Porosity')
sns.distplot( Sw , color="green", ax=axes[1, 0], axlabel='Water Saturation')
sns.distplot( Bo , color="teal", ax=axes[1, 1], axlabel='Formation Volume Factor (R_ft3/st_ft3)')
sns.distplot( WOC , color="lime", ax=axes[1, 2], axlabel='WATER OIL CONTACT (WOC)')
sns.distplot( COUNTER , color="brown", ax=axes[1, 3], axlabel='COUNTER')
plt.tight_layout()
plt.show()
# plt.savefig('Property_Distribution.png')   # to save the figure as a .PNG file in the working directory
```

```python
                 ##### Calculations For STOIIP & Recoverable Resources Calculations ##### 


GRV=Area*(Thickness)                                                    # GRV in ft^3  , GROSS ROCK VOLUME
NRV = GRV * NTG                                                       # NRV  in ft^3 , NET ROCK VOLUME
HCPV = NRV * POR * (1-Sw)                                             # HCPV  in ft^3 ,HYDROCARBON PORE VOLUME
OIIP = (HCPV / 1E6) / (5.615*Bo)                                      # OIIP in mmbbls ,STOCK TANK OIL INITIAL IN PLACE

# Statistics
print("The minimum OIIP = %.5f mmbbls" % (min(OIIP)))
print("The maximum OIIP = %.5f mmbbls" % (max(OIIP)))
print("The mean OIIP = %.5f mmbbls" % (statistics.mean(OIIP)))                     # Arithmetic mean (“average”) of data
print("The OIIP std dev = %.5f mmbbls" % (statistics.stdev(OIIP)))                 # Sample standard deviation of data
print("The median OIIP = %.5f mmbbls" % (statistics.median(OIIP)))                 # Median (middle value) of data
print("The median_grouped OIIP = %.5f mmbbls" % (statistics.median_grouped(OIIP))) # Median or 50th percentile of grouped data
print("P99 OIIP = %.5f mmbbls" % (np.percentile(OIIP, 1))) 
print("P90 OIIP = %.5f mmbbls" % (np.percentile(OIIP, 10))) 
print("P50 OIIP = %.5f mmbbls" % (np.percentile(OIIP, 50))) 
print("P10 OIIP = %.5f mmbbls" % (np.percentile(OIIP, 90))) 
print("P01 OIIP = %.5f mmbbls" % (np.percentile(OIIP, 99)))
```
